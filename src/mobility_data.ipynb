```python
import pandas as pd
import matplotlib.dates as dates
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import files
data_2021 = pd.read_csv('drive/MyDrive/Data/2021_PH_Region_Mobility_Report.csv')
data_2022 = pd.read_csv('drive/MyDrive/Data/2022_PH_Region_Mobility_Report.csv')
```

```python
# Merge data
merged_df = pd.merge(data_2021, data_2022, on='date', how='outer', suffixes=('_2021', '_2022'))
merged_df.head()
```

```python
# Convert the date column to datetime
merged_df['date'] = pd.to_datetime(merged_df['date'])

# Set the date column as the index
merged_df.set_index('date', inplace=True)
```

```python
col_names = {
    ('retail_and_recreation_percent_change_from_baseline_2021',
     'retail_and_recreation_percent_change_from_baseline_2022'): 'Retail and Recreation',
    ('grocery_and_pharmacy_percent_change_from_baseline_2021',
     'grocery_and_pharmacy_percent_change_from_baseline_2022'): 'Grocery and Pharmacy',
    ('parks_percent_change_from_baseline_2021',
     'parks_percent_change_from_baseline_2022'): 'Parks',
    ('transit_stations_percent_change_from_baseline_2021',
     'transit_stations_percent_change_from_baseline_2022'): 'Transit Stations',
    ('workplaces_percent_change_from_baseline_2021',
     'workplaces_percent_change_from_baseline_2022'): 'Workplaces',
    ('residential_percent_change_from_baseline_2021',
     'residential_percent_change_from_baseline_2022'): 'Residential',
}
```

```python
# Set aesthetics
sns.color_palette('colorblind', )
sns.set_style("darkgrid")

fig, ax = plt.subplots(figsize=(10, 8))

merged_df.index = pd.to_datetime(merged_df.index)
merged_df = merged_df[~merged_df.index.duplicated(keep='first')]  # Keep the first occurrence of duplicate dates

for cols, name in col_names.items():
    for col in cols:
        sns.lineplot(data=merged_df, x=merged_df.index, y=col, label=name, ax=ax)

    ax.set_title('Mobility Trends by Region')
    ax.set_xlabel('Date')
    ax.set_ylabel('Percentage Change')

# Customize the x-axis to show only month labels
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b %Y'))

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Set aesthetics
sns.set_style("darkgrid")

fig, ax = plt.subplots(figsize=(16, 8))

merged_df.index = pd.to_datetime(merged_df.index)
merged_df = merged_df[~merged_df.index.duplicated(keep='first')]

# Get a colorblind-friendly palette
colors = sns.color_palette("colorblind", n_colors=len(col_names))
colors[1] = (0.5, 0.5, 0.5)  # Set the second color to a lighter gray
colors[5] = 'brown'

# Iterate over categories and plot both years together
for i, ((col_2021, col_2022), name) in enumerate(col_names.items()):
    sns.lineplot(data=merged_df, x=merged_df.index, y=col_2021, label=name, ax=ax, color=colors[i])
    sns.lineplot(data=merged_df, x=merged_df.index, y=col_2022, ax=ax, color=colors[i])

ax.set_title('Mobility Trends by Region')
ax.set_xlabel('Date')
ax.set_ylabel('Percentage Change')
# Customize the x-axis to show only month labels
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b %Y'))

plt.axhline(y=0, color='black', linestyle='-',linewidth=0.5)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
